```python
# função de alta ordem -> função que recebe outra função como argumento
# 1 - uso de programação funcional
def quadrado(x):
    return x * x

def aplicar_funcao(funcao, lista):
    return [funcao(x) for x in lista]

r = aplicar_funcao(quadrado, lista=[2,3,4,5]) # como se fosse um map, mas no caso é de alta ordem

print(r) # mesmo uso do map, retornando uma lista nesse caso
```

```python
# dados de exemplo - dicionário
usuarios = [
    {"nome":"Anna", "idade": 20},
    {"nome":"Marcelo", "idade":30},
    {"nome":"Matheus", "idade": 20},
    {"nome":"Gabriel", "idade": 21}
]

# o map, inseri a função dentro de uma lista, ou qualquer objeto iterável, dentro dele há iterador
def extrair_nomes(lista_de_nomes):
    return list(map(lambda m:m['nome'], lista_de_nomes)) # solução bem sofisticada

print(extrair_nomes(usuarios))

# o filter, inseri uma função dentro de um determinado iterável e retorna apenas os verdadeiros
def filtrar_maiores_que(usuario, idade):
    return list(filter(lambda u:u['idade'] > idade, usuario))

# todos os usuários que tem mais de 20 anos
print(filtrar_maiores_que(usuarios, 20))

def calcular_a_media_de_idades(usuario):
    if not usuario:
        return
    # sum retorna o somátorio no python, já pega os iteravéis de todas as idades em um formato de lista
    total_de_notas = sum(map(lambda u:u['idade'], usuario))

    return total_de_notas / len(usuario)

print(calcular_a_media_de_idades(usuarios))
```

```python
from functools import lru_cache

# decorator do lru_cache, função dentro de uma função, no bloco de código
@lru_cache(maxsize=None)
def firbonnaci(x): # 4
    if x <= 1:
        return x    
    return firbonnaci(x - 1) + firbonnaci(x - 2)

for i in range(10):
    print(f"firbonacci - {i} = {firbonnaci(i)}")
```
