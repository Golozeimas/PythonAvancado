```python
from collections import Counter
import re

def encontra_urls(texto_com_url):
    regex_para_urls = r"https?://[a-zA-Z0-9.-]+\.[a-zA-Z]{2,}"
    urls = re.findall(regex_para_urls, texto_com_url)
    return urls

textos_com_url = """
Veja nosso site em https://exemplo.com para mais informações.
O repositório está em http://github.com
Falha ao acessar http://localhost:8080/api/teste.com.br ontem.
Link inválido: ftp://site.com
Outro link: https://sub.dominio.com/path?x=1#secao
"""

print(encontra_urls(textos_com_url))
```

```python
def validar_senha(senha):
    regex_para_senha = r"^(?=.*[a-z])(?=.*[A-Z])(?=.*\d).{8,}$"
    # pattern é a regra do negócio
    valida = re.match(regex_para_senha, senha)
    
    if valida:
        return True
    else:
        return False

senha_sem_uso_do_padrao = "Asdsads"

senha_com_uso_do_padrao = "Jubileu12345"

for i in range(1):
    negado = validar_senha(senha_sem_uso_do_padrao)
    valido = validar_senha(senha_com_uso_do_padrao)
    print(negado)
    print(valido)
```

```python
def conta_as_palavras(texto):
    # esse regex corta as palavras tanto por vírgula quanto por espaço
    palavras_cortadas  = re.split(r"\W+", texto)
    cortadas = len(palavras_cortadas) + 1
    for i in range(cortadas):
        print(f"Quantidade de palavras cortadas: {i}")

conta_as_palavras("ass,as,ad,te,as")
```
