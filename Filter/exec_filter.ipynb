Filter
- É usada para filtrar elementos de um iterável com base em uma função que pode ser True ou False
- O objetivo é permitir apenas os elementos que retornarem True

```python
def numeros_pares(x):
    return x % 2 == 0

lista_de_numeros = [2,3,4,5,6,7,8,9,10]

r = filter(numeros_pares, lista_de_numeros)

list(r)
```

```python

produtos = [
    {"nome":"Produto A","preco":50.0, "em_estoque":True},
    
    {"nome":"Produto B","preco":40.0, "em_estoque":False},

    {"nome":"Produto C","preco":20.0, "em_estoque":True},
    
    {"nome":"Produto D","preco":10.0, "em_estoque":False}
]

def verifica_produtos_em_estoque(produto):
    return produto["em_estoque"] 

# o filter já faz o loop automaticamente, não preciso de índice para acessar
# só precisamos do nome do atributo e a palavra return
produtos_disponiveis = filter(verifica_produtos_em_estoque, produtos)

list(produtos_disponiveis)
```

```python
transacoes = [100, -120, 200, -102, -10, 300]

def transacoes_acima_de_0(transacao):
    return transacao > 0

resultados = filter(transacoes_acima_de_0, transacoes)

list(resultados)
```
