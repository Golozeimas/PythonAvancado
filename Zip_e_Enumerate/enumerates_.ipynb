```python
# enumerate
lista_de_pessoas = ["matheus","anna","marcela","igor","lucas"]

for i, valor in enumerate(lista_de_pessoas):
    print(f"índice {i} - pessoa: {valor}")


numeros = [10,20,30,40,20]

numeros_com_chaves = {indice:valor 
                      for indice, valor in enumerate(numeros)} # faz um for direto dentro do dicionário
                                                               # e já coloca os dados disponíveis do enumerate
numeros_com_chaves

indices_pares = {indice:valor 
                 for indice, valor in enumerate(numeros) if indice % 2 == 0}
indices_pares
```

```python
# zip
nomes = ["matheus", "ana", "lucas"]
idades = [20, 40, 18]

for nome, idade in zip(nomes, idades):
    print(f"nome: {nome} e idade: {idade}")

pessoas = dict(zip(nomes, idades))
pessoas

pessoas_acima_de_19 = {nome:idade for nome, idade in zip(nomes, idades) if idade > 19} # uso do for direto no dicionário
                                                                                       # parecido com o list comprehesion

pessoas_acima_de_19
```
