```python
def criador_de_pares_infinitos():
    i = 0
    while True:
        i = i + 2
        yield i
    """ Crie um generator que produza números pares infinitamente, começando do 0.
Use next() para pegar apenas os 10 primeiros """
```

```python
gerador = criador_de_pares_infinitos()

for _ in range(20):
    print(next(gerador))
```

```python
# o python com yield vai devolver sempre um objeto do tipo 'iterável'
def leitura(arquivo):
    with open(arquivo,"r", encoding="utf-8") as arq:
        linha = arq.readlines()
        yield linha
```

```python
vendas = leitura("vendas.txt")

print(vendas) # retorna o objeto iteravel que precisa do next ou de um for para ser visualizado

for venda in vendas:
    print(venda)
```
