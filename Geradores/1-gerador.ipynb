```python
def numeros_impares(n):
    for i in range(1, n+1, 2):
        yield i
```

```python
for numero in numeros_impares(20):
    print(numero)
```

```python
def linhas_com_numeros_de_palavras(arquivo, numeros_de_palavras):
    with open(arquivo, "r", encoding="utf-8") as file:
        for linha in file:
            if len(linha.split()) == numeros_de_palavras:
                yield linha.strip()
```

```python
arquivo = "arquivo.txt"
numero_palavras = 3

for linha in linhas_com_numeros_de_palavras(arquivo, numero_palavras):
    print(linha)
```
