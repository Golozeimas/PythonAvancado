Map -
- Função que aplica uma função a todos os itens de um iterável
- maneira eficiente e concisa de transformar dados 

```python
lista_de_numeros = [2, 3 , 4, 5 ,6]

def dobrar_numeros(lista):
    return lista * 2

resultado = map(dobrar_numeros, lista_de_numeros)

print(list(resultado))
```

```python
def numeros_ao_quadrado(x):
    return x ** 2

resultado2 = map(numeros_ao_quadrado, lista_de_numeros)

list(resultado2)
```

```python
def celsius_para_fahrenheit(lista_com_celsius):
    return (lista_com_celsius * 9/5) + 32

lista_com_celsius = [40, 30, 20 , 10]

resultado3 = map(celsius_para_fahrenheit, lista_com_celsius)

list(resultado3) 
```

```python
def palavras_em_minusculos_e_sem_espacos(x):
    return x.strip().lower()

lista_de_palavras = ["Joao"," Matheus ", " cosmic "]

resultado4 = map(palavras_em_minusculos_e_sem_espacos, lista_de_palavras)

list(resultado4)
```

```python
lista_de_numeros2 = [40,50,20,30]

# usando lambda no map
r = map(lambda x:x*2, lista_de_numeros2)

list(r)
```
