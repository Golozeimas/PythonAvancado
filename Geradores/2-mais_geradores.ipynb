```python
# 1 - gerador de expressão

quadrados = (x **2 for x in range(10))

for quadrado in quadrados:
    print(quadrado)
```

```python
# 2 - Iterador infinito
def contador_infinito():
    i = 0
    while True:
        yield i # Guarda o valor final e só para quando terminar         
        i += 1

contador = contador_infinito()
for _ in range(20):
    print(next(contador))
```

```python
# 3 - coleta de valores com send()

def coletador():
    total = 0
    while True:
        valor = yield total
        if valor is not None:
            total += valor

soma = coletador()
next(soma) # inicializa o gerador
print(soma.send(10))
print(soma.send(5))
print(soma.send(15))
print(soma.send(20))
```

```python
# 4 - tratamento de exceções
def excecao():
    try:
        while True:
            valor = yield
            print(f"Valor recebido: {valor}")
    except ValueError:
        print("Valor inválido")

ex = excecao()
next(ex) # Inicializa o gerador
ex.send(20)
ex.throw(ValueError)
```

```python
def cascata_de_geradores(iterable):
    for i in iterable:
        yield i * 2
    
def adicionar_cinco(iterable):
    for i in iterable:
        yield i + 5

numeros = range(0,5) # por se tratar de um range usamos esse iterable para percorrer

resultado = adicionar_cinco(cascata_de_geradores(numeros)) # lógica de dentro pra fora da matemática
                                                           # e da programação

for r in resultado: # esse for faz o trabalho do next
    print(r)
```
