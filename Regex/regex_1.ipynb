```python
import re

# re.search -> é usado para encontrar padrões de acordo com o 
# pattern do regex, e uma string, retorna apenas o primeiro e é mais rápido
def explorar_regular_expressao(texto):
    padrao = r"\bPython\b" # a palavra "Python"
    if re.search(padrao,texto):
        print("O padrão foi encotrado, parabêns!")
    else:
        print("Não foi possível encontrar o padrão!")



explorar_regular_expressao("gosto de Python")
explorar_regular_expressao("Não gosto de python")
```
