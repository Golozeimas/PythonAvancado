```python
class MeuMeta(type): # outra convenção é o uso do type na metaclasse
    def __new__(cls, name, bases,attrs): # isso é convenção da metaclasse
        if "id" not in attrs:
            raise TypeError(f"A classe {name} deve ter atribuído 'id' ")
        return super().__new__(cls, name, bases, attrs)
```

```python
class MetaObrigatoria(metaclass=MeuMeta):
    # id = 1
    nome = "Matheus" # vai dar erro pois a classe MetaObrigatoria, deve passar um id, seguindo a metaclasse
```
