```python
import itertools
```

```python
# Exemplo 1 - range()

for i in range(10):
    print(i)
```

```python
# Exemplo 2 - iterar na listas
lista_de_frutas = ["Maçã", "Banana", "Mamão"]

for lista in lista_de_frutas:
    print(lista)
```

```python
# Exemplo 3 - Iterar dicionário

dicionario_de_itens = {
    "nome_do_item":"celular",
    "valor_do_item":2000,
    "qualidade_do_item":["boa","média", "ruim"]
}

for i, valor in dicionario_de_itens.items():
    print(f"item: {i}, o valor: {valor}" )
```

```python
# Exemplo 4 - enumerate()

for i, value in enumerate(lista_de_frutas):
    print(f"Indice: {i}, valor: {value}")
```

```python
# Exemplo 5 - Zip()

nomes = ["Ana", "Matheus", "Clara", "Jairo"]

idades = [20, 30, 19, 20]

for n, i in zip(nomes, idades):
    print(n,"idade:", i) # combino duas informações de listas diferentes e compacto isso
```

```python
# Exemplo 6 - itertools

for i in itertools.count(start=10, step=2):
    if i > 20:
        break
    print(i)
```
