Uma lista é iterável pois possui na sua composição o metódo especial __iter__, enquanto números não são

Obs: strings são iteráveis

```python
lista = [1,2,3,4,5]

numeros = 12345

string = "matheus"

print(hasattr(lista, '__iter__')) # é iterável

print(hasattr(numeros, "__iter__")) # não é iterável

print(hasattr(string, "__iter__")) # é iterável
```

Funcionamento do for, envidencia o uso do iterador:
O for por si só é um iterável visto que ele transforma listas, strings, dicionários e tuplas em uma lista iterável

A variável faz o papel de guardar o next() do for, exemplo: "for x in lista", a lista é transoformada em iterável com a função "iter()", o x guarda o valor 1 por 1 do "next()"

```python
def funcionamento_do_for(lista):
    lista = iter(lista)
    while True:
        try:
            l = next(lista)
            print(l)
        except: 
            StopIteration
            break

lista = [1,2,3,4,5,6]

funcionamento_do_for(lista)
```

Imagine que eu tenha uma lista gigantesca isso consome muita memória, vamos exemplificar com código:

```python
import sys

lista = list(range(10000)) # isso destroi a memória, caso esteja trabalhando em um
# ambiente mais limitado, e também destrói a performance do seu código

print("Bytes na memória:", sys.getsizeof(lista))
# esses 80000 bytes de memória podem parecer pouco até mas ao passar do tempo
# essa lista só tende a aumentar cada vez mais
```

```python
import time

def gera():
    lista = []
    for i in range(100): # -1 do range (100), adiciona do 0 ao 99
        lista.append(i)
        time.sleep(0.1)
    return lista

g = gera()

for v in g:
    print(v) # 10 segundos para essa simples ação
```

```python
# usando a função geradora para o mesmo problema

def funcao_geradora():
    for i in range(100):
        yield i
        time.sleep(0.1)

g = funcao_geradora()

for i in g:
    print(i)
```

```python
# usando a função geradora para o mesmo problema

def funcao_geradora():
    for i in range(100):
        yield i
        time.sleep(0.1)

g = funcao_geradora()

print(next(g))
print(next(g))
print(next(g))
print(next(g))

```

```python
# gerador pt 2
def gera():
    valor1 = "Gerou o valor 1"
    yield valor1
    valor2 = "Gerou o valor 2"
    yield valor2
    valor3 = "Gerou o valor 3"
    yield valor3

g2 = gera()

print(next(g2))
print(next(g2))
print(next(g2))
```

```python
# agora, pra exemplificar melhor ainda a diferença

l1 = [x for x in range(1000)] # list comphresion

l2 = (x for x in range(1000)) # gerador

print("Tipo de dado: ", type(l1))
print("Tipo de dado: ", type(l2))

print(sys.getsizeof(l1)) # enquanto temos 8856 bytes na memória
print(sys.getsizeof(l2)) # aqui temos apenas 200 bytes na memória, uma redução drástica
```
