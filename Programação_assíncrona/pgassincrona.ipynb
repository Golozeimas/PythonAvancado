```python
import asyncio

async def tarefa(nome, tempo):
    print(f"Tarefa {nome} iniciada!")
    await asyncio.sleep(tempo)
    print(f"Tarefa {nome} concluída!")

async def main():
    await asyncio.gather( # consegue iniciar as três ao mesmo tempo
        tarefa("A",2),
        tarefa("B",1),
        tarefa("C",3)
    )

try:
    asyncio.get_running_loop()
except RuntimeError:
    asyncio.run(main())

asyncio.ensure_future(main())
```
