O type hint surge para que nós programadores python não se embarace no próprio código, determinando o tipo de dado visualmente, ou aguardando o retorno de um determinado tipo de dado numa função

```python
def soma(a: int, b: int) -> int:
    return a + b

print(soma(1,2))
```

```python
from typing import List
class Estudante:
    def __init__(self, nome: str, notas: List[float]) -> None:
        self.nome = nome
        self.notas = notas
    
    def media(self) -> float:
        return sum(self.notas) / len(self.notas)
    

def alunos_acima_da_media(alunos: List[Estudante], media_minima: float) -> List[str]:
    return [aluno.nome for aluno in alunos if aluno.media() >= media_minima] # fazendo uma list comphrension

lista_de_estudantes = [
    Estudante("Alice", [10,3,5,4,3]), # usando o construtor, dá acesso a classe
    Estudante("Matheus", [9,7,9,10,8]),
    Estudante("Patricio", [10,6,8,9,10])
]

print(alunos_acima_da_media(lista_de_estudantes, 8)) # passando aqui no primeiro parâmetros 
                                                     # temos os mesmos atributos e métodos da classe 'Estudante'
```
